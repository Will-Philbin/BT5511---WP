```python
aligner = Align.PairwiseAligner()

aligner.mode = "global"

aligner.match_score = 2
aligner.mismatch_score = -1
aligner.open_gap_score = -5
aligner.extend_gap_score = -0.5

alignments = aligner.align (Avon4_WT, Avon4_Mut)[0]

best = alignments
print(best)
```

```python
aligned_Avon4_WT, aligned_Avon4_Mut = alignments
```

```python
print(aligned_Avon4_WT)
print(aligned_Avon4_Mut)
```

```python
subs = 0
insertions = 0
deletions = 0

for a, b in zip(str(aligned_Avon4_WT), str(aligned_Avon4_Mut)):

    if a == b:
        continue

    if a != "-" and b != "-":
        subs += 1

    elif a == "-" and b != "-":
        insertions += 1

    elif a != "-" and b == "-":
        deletions += 1

print("Substitutions:", subs)
print("Insertions:", insertions)
print("Deletions:",deletions)
```
