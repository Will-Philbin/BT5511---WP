```python
subs = 0
insertions = 0
deletions = 0

for a, b in zip(str(aligned_Anap7c1_WT), str(aligned_Anap7c1_Mut)):

    if a == b:
        continue

    if a != "-" and b != "-":
        subs += 1

    elif a == "-" and b != "-":
        insertions += 1

    elif a != "-" and b == "-":
        deletions += 1

print("Substitutions:", subs)
print("Insertions:", insertions)
print("Deletions:",deletions)
```
